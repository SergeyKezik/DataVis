```python
import pandas as pd
import numpy as np
import plotly
```

```python
init_data = pd.read_excel('init_data.xlsx')
```

```python
init_data = init_data.drop('Name (Chinese)', axis = 1)
init_data = init_data.drop('Entity owner (Chinese)', axis = 1)
init_data = init_data.drop('Parent entity (Chinese)', axis = 1)
```

```python
init_data = init_data.rename(columns = { 'Name (English)':'name', 'Region of Focus':'region_focus', 'Language':'language',
       'Entity owner (English)':'owner', 'Parent entity (English)':'parent',
       'X (Twitter) handle':'x_page', 'X (Twitter) URL':'x_url', 'X (Twitter) Follower #':'x_follower_n',
       'Facebook page':'facebook_page', 'Facebook URL':'facebook_url', 'Facebook Follower #':'facebook_follower_n',
       'Instragram page':'instagram_page', 'Instagram URL':'instagram_url', 'Instagram Follower #':'instagram_follower_n',
       'Threads account':'threads_page', 'Threads URL':'threads_url', 'Threads Follower #':'threads_follower_n',
       'YouTube account':'youtube_page', 'YouTube URL':'youtube_url', 'YouTube Subscriber #':'youtube_follower_n',
       'TikTok account':'tiktok_page', 'TikTok URL':'tiktok_url', 'TikTok Subscriber #':'tiktok_follower_n' })
```

```python
parent_summary = init_data["parent"].value_counts()
parent_summary
```

```python
circle_media = [{'id': 'all', 'datum': init_data['x_follower_n'].sum() + init_data['facebook_follower_n'].sum() \
                 + init_data['instagram_follower_n'].sum() + init_data['threads_follower_n'].sum() \
                 + init_data['youtube_follower_n'].sum() + init_data['tiktok_follower_n'].sum() , 'children': []}]
```

```python
for i in range(parent_summary.shape[0]):
    entity = parent_summary.index[i]
    sub_dataset = init_data.loc[init_data['parent'] == entity]
    datum = sub_dataset['x_follower_n'].sum() + sub_dataset['facebook_follower_n'].sum() \
                 + sub_dataset['instagram_follower_n'].sum() + sub_dataset['threads_follower_n'].sum() \
                 + sub_dataset['youtube_follower_n'].sum() + sub_dataset['tiktok_follower_n'].sum() 
    children = sub_dataset["owner"].value_counts()

    if (children.shape[0] > 1):
        this_dict = dict()
        this_dict['id'] = entity
        this_dict['datum'] = datum
        this_dict['children'] = []
        
        for j in range(children.shape[0]):
            sub_sub_dataset = sub_dataset.loc[sub_dataset['owner'] == children.index[j]]
            this_dict['children'].append({'id': children.index[j], 'datum': sub_sub_dataset['x_follower_n'].sum() + \
                                          sub_sub_dataset['facebook_follower_n'].sum() \
                 + sub_sub_dataset['instagram_follower_n'].sum() + sub_sub_dataset['threads_follower_n'].sum() \
                 + sub_sub_dataset['youtube_follower_n'].sum() + sub_sub_dataset['tiktok_follower_n'].sum()})

        circle_media[0]['children'].append(this_dict)
            
    else:
        circle_media[0]['children'].append({'id': entity, 'datum': datum})    
```

```python
# import the circlify library
import circlify

# Compute circle positions thanks to the circlify() function
circles = circlify.circlify(
    circle_media, 
    show_enclosure=False, 
    target_enclosure=circlify.Circle(x=0, y=0, r=1)
)
```

```python
# import libraries
import circlify
import matplotlib.pyplot as plt

# Create just a figure and only one subplot
fig, ax = plt.subplots(figsize=(14,14))

# Title
ax.set_title('Repartition of the world population')

# Remove axes
ax.axis('off')

# Find axis boundaries
lim = max(
    max(
        abs(circle.x) + circle.r,
        abs(circle.y) + circle.r,
    )
    for circle in circles
)
plt.xlim(-lim, lim)
plt.ylim(-lim, lim)

# Print circle the highest level (continents):
for circle in circles:
    if circle.level != 2:
      continue
    x, y, r = circle
    ax.add_patch( plt.Circle((x, y), r, alpha=0.5, linewidth=2, color="lightblue"))

# Print circle and labels for the highest level:
for circle in circles:
    if circle.level != 3:
      continue
    x, y, r = circle
    label = circle.ex["id"]
    ax.add_patch( plt.Circle((x, y), r, alpha=0.5, linewidth=2, color="#69b3a2"))
    #plt.annotate(label, (x,y ), ha='center', color="white")

#Print labels for the continents
for circle in circles:
    if circle.level != 2:
      continue
    x, y, r = circle
    label = circle.ex["id"]
    plt.annotate(label, (x,y ) ,va='center', ha='center', bbox=dict(facecolor='white', edgecolor='black', boxstyle='round', pad=.5))


 
```

```python
with open("subs.txt") as inf:
    colors = inf.readlines()
```

```python
for i in range(len(colors)):
    colors[i] = float(colors[i].replace(',\n', '').replace('[', '').replace(']', ''))
```

```python
from scipy.stats import percentileofscore

# Calculating percentiles for each data point
percentiles = [percentileofscore(colors, x) for x in colors]

# Map percentiles to color vectors
# Using a non-linear mapping (e.g., square root) to ensure even color distribution
color_vectors_percentile = [
    (int(255 * (np.sqrt(p) / 10)), int(255 * (1 - (np.sqrt(p) / 10))), 0)
    for p in percentiles
]
```

```python
start_color = (6, 49, 98)  # RGB for starting color
end_color = (104, 1, 31)   # RGB for ending color

color_vectors_percentile = [
    (
        int(start_color[0] + (end_color[0] - start_color[0]) * p/100),
        int(start_color[1] + (end_color[1] - start_color[1]) * p/100),
        int(start_color[2] + (end_color[2] - start_color[2]) * p/100)
    )
    for p in percentiles
]
```

```python
hex_list = []
for rgb in color_vectors_percentile:
    # Convert each RGB tuple to HEX and append to the list
    hex_color = "#{:02x}{:02x}{:02x}".format(*rgb)
    hex_list.append(hex_color)
```
